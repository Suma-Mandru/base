```python
import numpy as np
#creating arrays
arr=np.array([1,2,3,4,5])
print(arr)
a=np.zeros((3,3),dtype=int)
print(a)
b=np.ones((2,2),dtype=int)
print(b)
c=np.arange(10)
print(c)

#array manipulation
manp=arr.reshape(5,1)
print(manp)

sliced=arr[2:4]
print(sliced)

a1=np.array([3,5,7,9])
a2=np.array([1,4,6,8])
result=np.vstack(a1+a2)
print(result)
```

```python
a1=np.array([3,5,7,9])
a2=np.array([1,4,6,8])
result=np.stack(a1+a2)
print(result)
#split func
a=np.array([3,5,7,9,1,6])
sp=np.split(a,2)
print(sp)
#transpose
a=np.array([[3,5,7],[9,1,6]])
b=a.T
print(b)
```

```python
import numpy as np
arr1=np.array([1,2,3,4,5])
#broadcasting
s=arr1+3
print(s)
arr2=np.array([6,7,8,9,0])
c=arr1+arr2
print(c)
```

```python
#linear algebra
m1=np.array([[8,6],[4,3]])
m2=np.array([[1,9],[7,2]])
r=np.dot(m1,m2)
print(r)
print(m1)
print(m2)
```

```python
#calculating eigen values and eigen vectors
m1=np.array([[8,6],[4,3]])
m2=np.array([[1,9],[7,2]])
r=np.dot(m1,m2)
print(r)
d=np.linalg.eig(r)
print(d)
```

```python
#stastical operations
a=np.array([1,2,3,4,5])
mean=np.mean(a)
print(mean)
median=np.median(a)
print(median)
variance=np.var(a)
print(variance)
standard_deviation=np.std(a)
print(standard_deviation)
```

```python
a=[230,560,780,127,128]
b=[200,160,270,127,400]
years=[1,2,3,4]
profit_a=[(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b=[(b[i]-b[i-1]) for i in range(1,len(b))]
plt.subplot(2,1,2)
plt.plot(years,profit_a,color='hotpink',linewidth='3',label='CompanyA',marker='>',ms='15',mec='k')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label='CompanyB',marker='H')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the anscombe dataset
anscombe = sns.load_dataset("anscombe")

# Display the first few rows of the dataset to understand its structure
print(anscombe.head())

# Plot the four datasets in the anscombe dataset
sns.set(style="ticks")
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=anscombe,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset("iris")
print(iris)
#compute correlation matrix
correlation_matrix=iris.corr()
#create a heatmap of the correlation matrix
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```
